## 虚拟环境

创建独立于系统的 Python 运行环境和煲管理环境，针对每个项目创建独立的虚拟环境。

- Python 3 之后自带的 venv 模块：[venv](https://docs.python.org/3/library/venv.html)
- virtualenv：[virtualenv](https://virtualenv.pypa.io/en/latest/)，[virtualenvwrapper](https://virtualenvwrapper.readthedocs.io/en/latest/)
- pipenv：[pipenv](https://github.com/pypa/pipenv)
- Anaconda：[Anaconda](https://docs.continuum.io/)

### venv 模块

```shell
# 创建
python3 -m venv project-env

# 进入虚拟环境
cd project-env

# 激活
source bin/activite

# 退出
deactivate
```

### virtualenv

```shell
# 安装系统目录（root权限）
pip install virtualenv

# 安装用户目录
pip install virtualenv --user
```

> 注意：在 `~/.bashrc` 中新增 `export PATH="$HOME.local//bin:$PATH"`

```shell
virtualenv project-env
```

**安装 virtualenvwrapper**：`pip install virtualenvwrapper`

安装后，在用户目录下的 `.bashrc` 文件添加内容

```shell
export WORKON_HOME=$HOME/.virtualenvs
source /usr/local/bin/virtualenvwrapper.sh

# 使用 --user 安装，使用一下方式
source $HOME/.local/bin/virtualenvwrapper.sh

# 创建
mkvirtualenv project-env

# 激活
workon project-env
```

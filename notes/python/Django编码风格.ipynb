3. choices 字段用法

带有 `choices` 参数字段，choices 的定义需要大写。

```python
class ModelName(models.Model):
    DIRECTION_UP = 'U'
    DIRECTION_DOWN = 'D'
    DIRCTION_CHOICES = (
        (DIRECTION_UP, 'Up'),
        (DIRECTION_DOWN, 'Down'),
    )
    direction = models.CharField(max_length=10, choices=DIRECTION_CHOICES);
```

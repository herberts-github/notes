## while 循环操作

`for` 循环用于针对集合中的每个元素执行一个代码块，而 `while` 循环则不断运行，直到指定条件不满足为止.

```python
current_number = 1
while current_number <= 5:
    print(current_number)
    current_number += 1
```

### 选择何时退出

使用 `while` 循环让程序不断运行，定义提示信息，输入退出值停止程序运行。

```python
prompt = '\n输入任意内容，输入 quit 退出运行：'
message = ''

while message != 'quit':
    message = input(prompt)
    # if 检查仅在消息不是退出值时打印
    if message != 'quit':
        print(message)
```

### 标志

在要求很多条件满足才能继续运行的程序中，定义用于判断整个程序是否处于活动状态的变量。

```python
prompt = '\n输入任意内容，输入 quit 退出运行：'
message = ''

active = True  # 用于判断陈旭是否继续运行
while active:
    message = input(prompt)

    if message == 'quit':
        active = False
    else:
        print(message)
```

### `break` 退出循环

`break` 用于控制程序流程，可用来控制那些代码执行、不执行，从而让程序按要求执行。

```python
prompt = '输入您喜欢的城市：'

while True:
    city = input(prompt)

    if city == 'quit':
        break # 当遇到 break ，将跳出循环
    else:
        print(f'城市：{ city }')
```

### `continue` 跳过循环

根据条件测试结果决定是否继续执行循环，返回循环开头。

```python
current_number = 0

while current_number < 10:
    current_number += 1
    if current_number % 2 == 0:
        continue
    print(current_number)
```

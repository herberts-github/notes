## 条件测试

每条 `if` 语句的核心都是一个值为 True 或 False 的表达式，成为条件测试。<br/>
Python 根据条件测试的值为 True 或 False 来决定执行 `if` 语句中的代码

### 检查是否相等

将变量的当前值和特定值进行比较
检查变量的值是否与特定值相等

````python
car = 'bmw'
# 使用 == 判断是否相等
car == 'bmw'
````

### 检查是否相等时忽略大小写

Python 中检查是否相等时区分大小写

可使用 `lower` 将变量值转换全小写，再进行比较<br/>
同理，`upper` 可将变量值转换全大写

````python
car = 'Audi'
# car.lower() == 'audi'

car.upper() == 'AUDI'
````

### 检查是否不相等

判断两个值是否不相等，可使用惊叹号和等号（!=）

````python
1 != 2
````

### 数值比较

条件语句可包含各种数据比较，例：小于、小于等于、大于、大于等于 等等

````python
1 < 2, 1 <= 2, 1 > 2, 1 >= 2
````

### 检查多个条件

可同时检查多个条件，可通过 `and`  和 `or` 将两个条件语句进行判断

````python
# and：检查是否两个条件都为 True，使用 and 将两个测试合二为一
one = 1
two = 2
(one != 1) and (two == 2)  # False
print((one == 1) and (two == 2))

# or：只要至少一个条件满足，便通过测试
print((one != 1) or (two == 2))
````

### 检查特定值是否包含列表中

确定特定值是否包含在列表中，可通过 `in` 或 `not in` 进行判断

````python
temp = [1, 2, 3]
print(f'列表包含 1：{1 in temp}')
print(f'列表包含 4：{4 in temp}')
````

## 条件语句

### if 语句

根据条件结果，进行判断布尔值执行操作

```python
if conditional_test:
    do something
```

### if-else 语句

条件测试通过时执行操作，在没有通过执行另一操作

```python
if conditional_test:
    do something
else:
    do something
```

### if-elif-else 语句

检查超过两个的情况下，可使用 if-elif-else 结构中的代码块<br/>
依次检查每个条件测试，直到遇到通过的条件测试，将紧跟条件后面的代码，并跳过余下测试

```python
if conditional_test:
    do_something
elif conditional_test:
    do_something
else:
    do_something
```

> **注意：同时可使用任意数量的 elif 代码块**

````python
nums = 7

if nums <= 21:
    print(f'小于或等于21：{nums}')
elif nums <= 79:
    print(f'小于或等于79：{nums}')
elif nums <= 92:
    print(f'小于或等于92：{nums}')
else:
    print('没有合适的条件判断')
````

### 省略 else 代码块

Python 并不要求 if-elif 结构后有 else 代码块

`else` 只要不满足任何 `if` 或 `elif` 中的条件测试，其中的代码就会执行

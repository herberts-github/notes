## for 循环操作

for 循环时，给依次与列表中每个值相关联的临时变量指定任意名称

```python
for num in nums:
    ```

### for 循环执行更多操作

for 循环中，可对每个元素执行任何操作

````python
names = ['abay', 'ben', 'coco']
for name in names:
    print(f'Hello, {name}')

# 在循环结束后，可提供总结性输出或接着执行程序必须完成的其他任务
print('Goodbye everyone!')
````

## 避免缩进错误

Python 根据缩进判断代码行与前一个代码行的关系，通过使用缩进让代码更易读

在较长 Python 程序中，将看到缩进程度各不相同的代码块，从而对程序的组织结构有大致认识

### 忘记缩进

````python
names = ['abay', 'ben', 'coco']
for name in names:
    print(f'Hello, {name}')
````

通常，**将紧跟在 for 语句后面的代码行缩进，可消除缩进错误**

### 忘记缩进额外的代码行

循环能够运行且不会报告错误，但结果可能出人意料

````python
names = ['abay', 'ben', 'coco']
for name in names:
    print(f'Hello, {name.title()}')
print(f"{name.title()}, It's a nice day today.")
````

从语法上看，Python代码是合法的，但由于存在逻辑错误，结果并不符合预期，<br/>
预期某项操作将每个元素执行一次，但只执行了一次，**请确定需要将一行还是多行代码缩进**

### 不必要的缩进

有时无须缩进，并非循环的组成部分，因此 Python 将指出这种错误

````python
message = 'hello world!'
print(message)
````

**避免意外缩进错误，只缩进需要缩进的代码块**

### 遗漏冒号

for 语句末尾冒号（:）告诉 Python，下一行是循环的第一行

````python
names = ['abay', 'ben', 'coco']
for name in names
    print(f'Hello, {name.title()}')
````

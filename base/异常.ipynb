## 异常

Python 使用称为**异常**的特殊对象来管理程序执行期间发生的错误。

使用 `try-except` 代码块处理的，让 Python 执行指定操作，同时告诉 Python 发生异常时怎么办。

即便出现异常，程序继续运行；显示编写友好的错误消息而不是 `traceback`。

### ZeroDivisionError

```python
print(5 / 0)
```

Python 无法按要求执行时，创建异常对象，
同时停止运行程序，并指出引发哪种异常，根据信息对程序进行修改。

### try-except

将导致错误的代码放在 `try` 代码块，成功跳过 `except`，导致错误将执行 `except` 代码块中的代码。

```python
try:
    print(5 / 0)
except ZeroDivisionError:
    print('不能除以0!')
```

### try-except-else

Python 尝试执行 `try` 代码块中的代码，只有可能引发异常的代码才需要放在 `try` 中。
有些仅在 `try` 代码块成功执行时需要运行代码放在 `else` 中。
`except` 代码块告诉 Python 引发指定异常执行的操作。

```python
print('除法运算：（输入 q 退出）')

while True:
    first_number = input('\n1:')
    if first_number == 'q':
        break
    second_number = input('2:')
    if second_number == 'q':
        break
    try:
        answer = int(first_number) / int(second_number)
    except ZeroDivisionError:
        print('不能除以 0 !')
    else:
        print(answer)
```

> 注意：
>
> 1. 编写经过详尽测试的代码不容易出现内部错误：语法或逻辑错误
>
> 2. 依赖外部因素可能出现异常：用户输入、存在指定的文件、网络链接
>
> 根据情况判断该在程序的什么地方包含异常处理块，以及出现错误时该向用户提供多少相关信息。

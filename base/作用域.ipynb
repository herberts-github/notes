## 作用域

变量可将其视作指向值的名称，执行赋值语句 x = 1 后，名称 x 指向值 1

```python
def foo(): x = 42  # local


x = 1  # global
print(x)
foo()
print(x)
```

赋值语句 `x = 42` 在内部作用域（**局部命名空间**）中执行，
不影响外部（**全局**）作用域内的 `x`

参数类似于**局部变量**，因此参数与全局变量同名不会造成影响

```python
def output(x): print(x)


x = 1
y = 2
output(y)
```

### global

**重新关联**全局变量（使其指向新值）

在函数内部给变量赋值时，该变量默认为 局部变量，可使用 `global` 明确指定全局变量

```python
x = 1


def change_global():
    global x
    x = x + 1


change_global()
print(x)
```

### 作用域嵌套

函数可嵌套，将一个函数放在另一个函数内

```python
def foo():
    def bar():
        print('hello world!')

    bar()


foo()
```

使用一个函数创建另一个函数

函数位于另一个函数内，且外面的函数**返回里面的函数**，返回一个函数，而不是调用（返回的函数能够访问其定义所在的作用域）

每当外部函数被调用时，都将重新定义内部的函数，而外部变量值也可能不同

**可在内部函数中访问来自外部局部作用域的变量**

```python
def multiplier(factor):
    # 存储其作用域函数称为 闭包
    def multiplyByFactor(number):
        return number * factor

    return multiplyByFactor


double = multiplier(2)
double(5)
# 10

triple = multiplier(3)
triple(3)
# 9

multiplier(5)(4)
# 20
```

> **注意：不能给外部作用域的变量赋值，但可使用关键字 `nonlocal` 使外部作用域（非全局作用域）内的变量赋值**

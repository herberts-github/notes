## 元组

- 列表：用于存储程序运行期间可变化的数据集，可变
- 元组：系列不可修改的元素（项目配置、常量），不可变

### 定义元组

使用圆括号标识，可使用索引访问其元素

```python
tuples = (79, 21)
print(tuples[0])
print(tuples[1])

# 因元组不能操作元素，修改元组将导致类型错误消息
tupels[0] = 92
```

### 遍历元素

使用 for 循环遍历元组所有值

```python
tuples = (79, 21, 92, 90)

for t in tuples:
    print(t)
```

### 修改元组变量

可通过给存储元组的变量赋值来修改元组的元素

```python
tuples = (79, 21, 92, 90)
tuples = (1, 2, 3, 4)
print(tuples)
```

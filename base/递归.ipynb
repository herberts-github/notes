## 递归

函数可调用自己，意味这引用（调用）函数自身

```python
def func():
    return func()
```

理论上，程序将不断运行，每次调用函数时，都将消耗一些内存，

当函数调用次数一定程度（且之前函数调用未返回）后，将耗尽所有内存空间，导致程序终止并显示错误消息（超过最大递归深度）

**无穷递归（无限循环）通常包含：**

- **基线条件**（针对最小的问题）：满足这种条件时函数将直接返回一个值
- **递归条件**：包含一个或多个调用，旨在解决**问题的一部分**

> 通过将问题分解为较小部分，可避免递归无穷执行（问题终将被分解成基线条件可解决的最小问题）
>
> 每次调用函数时，为此创建新的命名空间，意味着函数调用自身时，是两个不同的函数
>
> 可将此视作为两个属于相同物种的动物彼此交流

### 案例：阶乘和幂

- 阶乘

    计算数字 n 的阶乘

    n 的阶乘为 `n * (n - 1) * (n - 2) * ... * 1`

````python
# 循环
def factorial(n):
    # 将 result 设为 n
    result = n
    # 将其依次乘以 1 到 n-1 的每个数字
    for i in range(1, n):
        result *= i
    return result


# 1 的阶乘为 1
# 对于大于 1 的数字 n，其阶乘为 n-1 的阶乘再乘以 n


# 函数
def factorial(n):
    if n == 1:
        return 1
    else:
        return n * factorial(n - 1)
````

> 函数调用 `factorial(n)` 和 `factorial(n - 1)` 是不同的实体

- 计算幂

    内置函数 `pow` 和 运算符 `**`

    power(x, n)：x 的 n 次幂，将数字 x 自乘 n - 1 次的结果，n 个 x 相乘的结果

````python
def power(x, n):
    result = 1
    for i in range(n):
        result *= x
    return result


# 对于任何数字 x，power(x, 0) 都为 1
# n>0 时，power(x, n) 为 power(x, n-1) 与 x 的乘积

def power(x, n):
    if n == 0:
        return 1
    else:
        return x * power(x, n - 1)
````
